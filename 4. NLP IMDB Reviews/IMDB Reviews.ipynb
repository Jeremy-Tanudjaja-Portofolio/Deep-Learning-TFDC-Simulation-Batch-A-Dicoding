Getting the Data

```python
data, info = tfds.load('imdb_reviews', with_info = True, as_supervised =True, split=['train','test'])
```

```python
train_data, test_data = data
```

Preparing The Data

```python
# Without Decoding
train_x = []
train_y = []
for sentence, label in train_data:
  train_x.append(sentence)
  train_y.append(label)
```

```python
train_x[0]
```

```python
train_y[0]
```

```python
# Without Decoding but with numpy
train_x = []
train_y = []
for sentence, label in train_data:
  train_x.append(sentence.numpy())
  train_y.append(label.numpy())
```

```python
train_x[0]
```

```python
train_y[0]
```

```python
# with decoding but without numpy
train_x = []
train_y = []
for sentence, label in train_data:
  train_x.append(sentence.decode('utf8'))
  train_y.append(label)
# as you can see that it gives us an error if we don't change the type to numpy
```

```python
# With Decoding and numpy
train_x = []
train_y = []
for sentence, label in train_data:
  train_x.append(sentence.numpy().decode('utf8'))
  train_y.append(label.numpy())
```

```python
train_x[0]
```

```python
train_y[0]
```

```python
# For the test data
test_x = []
test_y = []
for sentence, label in test_data:
  test_x.append(sentence.numpy().decode('utf8'))
  test_y.append(label.numpy())
```

Now we wrap these the training and test data with an array because tensorflow expects the dimension to be
>[[data1],[data2],[data3]]
><br> and not
><br>[data1, data2, data3]

```python
train_x = np.array(train_x)
train_y = np.array(train_y)
```

```python
test_x = np.array(test_x)
test_y = np.array(test_y)
```

```python
train_x.shape
```

Parameters for the tokenizer and neural network

```python
vocab_size = 10000
embedding_dim = 16
max_length = 120
trunc_type = 'post'
oov_token = '<OOV>'
```

Create the tokenizer

```python
tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words=vocab_size, oov_token=oov_token)
```

```python
tokenizer.fit_on_texts(train_x)
```

Make the training sentences and testing sentences to be the tokenized version of it

```python
training_sentences = tokenizer.texts_to_sequences(train_x)
testing_sentences = tokenizer.texts_to_sequences(test_x)
```

Truncate the texts

```python
training_sentences = tf.keras.preprocessing.sequence.pad_sequences(training_sentences, truncating = trunc_type, maxlen = max_length)
testing_sentences = tf.keras.preprocessing.sequence.pad_sequences(testing_sentences, truncating = trunc_type, maxlen = max_length)
```

Create The Model

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, embedding_dim, input_length = max_length),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64,return_sequences=True)),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(32)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(loss='binary_crossentropy', metrics=['accuracy'], optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001))
```

```python
model.fit(training_sentences, train_y, epochs=10, validation_data=(testing_sentences, test_y))
```
